````python
# Chart 1: Food group consumption trends for Canada
line_chart_canada = alt.Chart(
    food_groups_time_series.query('Country == "Canada" and Year >= 2000 and Year <= 2013')
).mark_line(point=True).encode(
    x=alt.X('Year:T', title='Year'),
    y=alt.Y('Calories:Q', title='Calories (per capita)'),
    color=alt.Color('Food_group:N', title='Food Group'),
    tooltip=['Year:T', 'Food_group:N', 'Calories:Q']
).properties(
    width=500,
    height=250,
    title='Canada: Food Group Trends (2000-2013)'
)

line_chart_canada
````

````python
# Chart 2: Proportional food group consumption by country (normalized stacked bar)
bar_chart_proportions = alt.Chart(
    calories_food_groups_tidy.query('Country in ["Canada", "China", "India"]')
).mark_bar().encode(
    y=alt.Y('Country:N', title=''),
    x=alt.X('Calories:Q', title='Proportion of Calories', stack='normalize'),
    color=alt.Color('Food_group:N', title='Food Group'),
    tooltip=['Country:N', 'Food_group:N', 'Calories:Q']
).properties(
    width=350,
    height=250,
    title='Food Group Proportions by Country (2013)'
)

bar_chart_proportions
````

````python
# Step 2: Create the pivot table with aggregation to handle any duplicates
canada_wide = food_groups_time_series.query(
    'Country == "Canada" and Year >= 2000 and Year <= 2013'
).pivot_table(
    index=['Year'],
    columns='Food_group',
    values='Calories',
    aggfunc='sum'  # ADD THIS - handles any duplicate Year/Food_group combinations
).reset_index()

# Then create the chart
scatter_canada = alt.Chart(canada_wide).mark_circle(size=200).encode(
    x=alt.X('Cereals and Grains:Q', title='Cereals (Calories)', scale=alt.Scale(zero=False)),
    y=alt.Y('Meat:Q', title='Meat (Calories)', scale=alt.Scale(zero=False)),
    color=alt.Color('Year:T', title='Year', scale=alt.Scale(scheme='viridis')),
    tooltip=['Year:T', 'Cereals and Grains:Q', 'Meat:Q']
).properties(
    width=350,
    height=250,
    title='Canada: Cereals vs Meat Consumption (2000-2013)'
)

scatter_canada
````
